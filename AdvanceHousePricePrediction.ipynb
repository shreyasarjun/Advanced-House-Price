# Annova Test

```python
catCols=X_train.select_dtypes(include='object').columns.tolist()
len(catCols)
```

```python
X_train[catCols].info()
```

```python
X_train.shape,X_test.shape
```

```python
import statsmodels.stats.multicomp as multi
```

```python
DropingCols=set()
SelectedCols=set()
for i in catCols:
    if X_train[i].nunique()<3:
        print('simple annova:',i)
        continue
        
    mc1=multi.MultiComparison(X_train['SalePrice'],X_train[i])
    res1=mc1.tukeyhsd()
    reject=len(res1.reject[res1.reject==True])/len(res1.reject)
    
    if reject>=.5:
        SelectedCols.add(i)
    else:
        DropingCols.add(i)
```

```python
DropingCols
```

```python
SelectedCols
```

```python
X_train.drop(DropingCols, axis=1, inplace=True)
X_test.drop(DropingCols, axis=1, inplace=True)
X_train.shape, X_test.shape
```
